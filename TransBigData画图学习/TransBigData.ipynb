```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import transbigdata as tbd
```

```python
df_poi = pd.read_csv("hzpoi.txt",header=None,names=['ID','name','type','lng','lat'],dtype={'type':object},sep=';',encoding = 'utf-8')
df_poi.head(2)
```

```python
#设定显示范围
lon1 = min(df_poi['lng'])
lat1 = min(df_poi['lat'])
lon2 = max(df_poi['lng'])
lat2 = max(df_poi['lat'])
bounds = [lon1,lat1,lon2,lat2]
fig = plt.figure(1,(8,8),dpi = 150)  
ax = plt.subplot(111)
plt.sca(ax)
plt.scatter(df_poi['lng'],df_poi['lat'],c = "blue" ,s= 0.05,alpha=0.5)
tbd.plot_map(plt,bounds,zoom = 12,style = 4)  # 通过style参数进行调整
tbd.plotscale(ax,bounds = bounds,textsize = 10,compasssize = 1,accuracy = 'auto',rect = [0.06,0.03])
plt.show()
```
